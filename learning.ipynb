```python
class Encode_Sequnece():
    def __init__(self, sequence, encoding_dict):
        self.sequence = sequence
        self.encoding_dict = encoding_dict
        self.matrix = []
        self.comp_matrix = []
        self.scorekeep = {}
        self.ND_matrix = {}
        self.aligned_list = []

    def encode_str(self):
        l = []
        for charecter in self.sequence:
            l.append(self.encoding_dict[charecter])
        self.matrix.append(l)
        pprint(self.matrix)

    def differentiate(self, number_diffrenciation):
        for x in range(number_diffrenciation):
            l = []
            for y in range(0,len(self.matrix[x])-1):
                l.append(self.matrix[x][y]-self.matrix[x][y+1])
            self.matrix.append(l)
        pprint(self.matrix)

    def difference(self, object):
        self_matrix = self.matrix
        object_matrix = object.matrix
        self.comp_matrix = object_matrix
        # pprint("self",np.array(self_matrix))
        # pprint('obj',np.array(object_matrix))
        
        for  z in range(len(object_matrix)):
            print("________________________")
            
            diff_iteration = len(object_matrix)-z-1
            hold = []
            self.match_seq(np.array(self_matrix[diff_iteration]),object_matrix[diff_iteration],diff_iteration)
            
             

    def check_real(self,tup):
        print("flag",tup)
        seq = self.matrix[0][tup[0]:tup[0]+tup[2]+1]
        sub_seq = self.comp_matrix[0][tup[1]:tup[1]+tup[2]+1]
        print(seq,sub_seq)
        hold = np.subtract(np.array(seq),np.array(sub_seq))
        if len(np.where(hold!=0)[0])==0:
            self.scorekeep[tup] = {'score':tup[2]+1,'link':[tup]}
            

    def match_seq(self,arr1,arr2,diff_iteration):
        for y in range(len(arr2)):
            diffrence = (arr1-arr2[y])
            print('diffrence',diffrence)
            hit_list = np.where(diffrence == 0)
            #print("row:",hit_list[0])
            for x in hit_list[0]:
                self.check_real((x,y,diff_iteration))



    def is_overlap(self,range1, range2):
    # Check if range1 is entirely to the left or entirely to the right of range2
        if range1[1] < range2[0] or range1[0] > range2[1]:
            return False
        else:
            return True
  

    def link(self,tup1,tup2):
        (x1,y1,d1)=tup1
        (x2,y2,d2)=tup2
        
        # print('[X]major sub seq for 1st tuple',[tup1[0],tup1[0]+tup1[2]],'[Y]minor sub seq for 1st tuple',[tup1[1],tup1[1]+tup1[2]])
        # print('[X]major sub seq for 2nd tuple',[tup2[0],tup2[0]+tup2[2]],'[Y]minor sub seq for 2st tuple',[tup2[1],tup2[1]+tup2[2]])
        if((x1+d1<x2) and (y1+d1<y2)) or (x1>x2+d2 and y1>y2+d2):
            return True
        else:
            return False

    def align2(self):
        new_lists = []
        for x in self.scorekeep.copy():
            print("_________")
            print(x)
            h = len(self.scorekeep)
            print(len(self.scorekeep))
            hold_copy = self.scorekeep.copy()
            keys = list(hold_copy.keys())
            print(keys)
            keys = new_lists + keys
            for before_x in keys:
                if x == before_x:
                    break
                check_cond = True
                print(before_x)
                for link_element in self.scorekeep[before_x]['link']:
                    check_cond = self.link(x,link_element)
                    print(x,link_element,check_cond)
                    if check_cond == False:
                        print("fail")
                        break
                if check_cond == True :
                    print("Pass")
                    (x1,y1,d1,len1)=before_x
                    (x2,y2,d2,len2)=x
                    link = self.scorekeep[before_x]['link'].copy()
                    print(link)
                    link.append(x)
                    print(link)
                    self.scorekeep[(x1,y1,d1,len1+1)] = { 'score':self.scorekeep[before_x]['score']+self.scorekeep[x]['score']
                    ,'link':link}
                    new_lists.append((x1,y1,d1,len1+1))
                    print('added:',(x1,y1,d1,len1+1),self.scorekeep[(x1,y1,d1,len1+1)])


    def align(self):
        list_cords = list(self.scorekeep.values())
        old_list = list_cords.copy()
        for x in range(len(list_cords.copy())):
            print(old_list[x]["link"][0])
            semi_old_list = list_cords.copy()
            for y in range(len(semi_old_list[:x])):
                for list_vals in semi_old_list[y]["link"]:
                    check_cond = self.link(old_list[x]["link"][0],list_vals)
                    # print(old_list[x]["link"][0],list_vals,check_cond)
                    if check_cond == False:
                        print("fail")
                        break
                if check_cond == True :
                    print("Pass")
                    link = semi_old_list[y]["link"].copy()
                    # print(link)
                    link.append(old_list[x]["link"][0])
                    # print(link)
                    print({ 'score':old_list[x]['score']+semi_old_list[y]['score'],'link':link})
                    list_cords.insert(y, { 'score':old_list[x]['score']+semi_old_list[y]['score'],'link':link})
                    print(list_cords)
                    # self.scorekeep[(x1,y1,d1,len1+1)] = { 'score':self.scorekeep[before_x]['score']+self.scorekeep[x]['score']
                    # ,'link':link}
                    # new_lists.append((x1,y1,d1,len1+1))
                    # print('added:',(x1,y1,d1,len1+1),self.scorekeep[(x1,y1,d1,len1+1)])
        self.aligned_list = list_cords
                    
                
            


    def show(self,list_tup):
        for tup in list_tup:
            
            print(tup,self.sequence[tup[0]:tup[0]+tup[2]+1])
        
                    
            
        
                
                
        
```
